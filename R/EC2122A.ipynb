Estimate the variance step by step. First estimate the mean, then take the differences, square them, then take the mean of the square differences. Then, corroborate that you get the same result as the funcion `var`


$$
\sigma^2=\frac{1}{(N-1)}∑_i^N(x_i-\mu)^2
$$


````r
# Estimating the variance

mean_pop <- mean(data$population) # Estimating the mean
diffs <- data$population - mean_pop # Differences
diffs_squared <- diffs^2 # Square differences
sigma_2 <- sum(diffs_squared)/(length(diffs_squared)-1) #Estimating the variance

sigma_2

var(data$population)
````
